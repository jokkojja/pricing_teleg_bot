```python
import pandas as pd
import os
import json
```

```python
path = '/Users/jokkojja/Desktop/pet_projects/pricing_teleg_bot/cian_parser/data/'
```

```python

files = [file for file in os.listdir(path) if '.json' in file]
```

```python
df = pd.DataFrame()
for file in files:
    file = path + file
    with open(file) as prop_file:
        prop = json.load(prop_file)['properties']

    # converting json dataset from dictionary to dataframe
    df = df.append(pd.DataFrame.from_dict(prop))
df.index = range(len(df))    
```

```python
def extract_coords(dic):
    return pd.Series(dic)
```

```python
df[['lat', 'lng']] = df.apply(lambda x: extract_coords(x['coordinates']), axis = 1)
df = df.drop('coordinates', axis = 1)
```

```python
df = df.rename({'lat': 'latitude', 'lng': 'longitude'})
```
